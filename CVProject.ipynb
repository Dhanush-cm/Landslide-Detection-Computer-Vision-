```python
from ultralytics import YOLO
import gradio as gr
import cv2
import numpy as np
```

```python
model = YOLO("/content/ultralytics/runs/detect/train/weights/best.pt")
```

```python
def detect_landslide(image):
    # Run prediction
    results = model.predict(source=image, imgsz=640)
    # Convert results first image with bounding boxes to np.ndarray
    annotated_image = results[0].plot()
    return annotated_image
```

```python
iface = gr.Interface(
    fn=detect_landslide,
    inputs=gr.Image(type="numpy"),
    outputs=gr.Image(type="numpy"),
    title="Landslide Detection with YOLOv11",
    description="Upload an image to detect landslides using the trained YOLOv11 model."
)
```

```python
iface.launch(share = True)
```
